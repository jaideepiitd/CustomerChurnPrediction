```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/CC.csv')
```

```python
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
```

```python
data.info()
```

```python
data.describe()
```

```python
data['TotalCharges'].isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer1=SimpleImputer(strategy='mean')
imputer2=SimpleImputer(strategy='most_frequent')

```

```python
data['TotalCharges'].isnull().sum()
```

```python
X=data.drop(columns=['customerID','Churn'], axis=1)
X
```

```python
Y=data['Churn']
Y
```

```python
num_cols=['tenure','MonthlyCharges','TotalCharges']
```

```python
import numpy as np
X[num_cols] = X[num_cols].replace(' ', np.nan)
```

```python
cat_cols=[col for col in X.columns if col not in num_cols]
cat_cols
```

```python
[X[i].unique() for i in cat_cols]
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
```

```python
from sklearn.pipeline import make_pipeline
pipe1=make_pipeline(imputer1, scaler)
pipe2=make_pipeline(imputer2, ohe)
```

```python
from sklearn.compose import make_column_transformer
transformer=make_column_transformer(
    (pipe1, num_cols),
    (pipe2, cat_cols),
    remainder='drop',
    n_jobs=-1
)
```

```python
X_new=transformer.fit_transform(X)
X_new
```

```python
Y.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y=le.fit_transform(Y)
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_new, Y, test_size=0.25, random_state=40)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
xgb=GradientBoostingClassifier(n_estimators=300,
                               learning_rate=0.1,
                               max_depth=5,
                               min_samples_leaf=5,
                               min_samples_split=5)
```

```python
model=xgb.fit(X_new,Y)
y_pred=model.predict(X_test)
y_pred
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```
